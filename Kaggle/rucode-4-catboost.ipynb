### Finding catboost parameters

```python
# class weights
from sklearn.utils.class_weight import compute_class_weight
classes = np.unique(y_train.values)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train.values.reshape(-1))
class_weights = dict(zip(classes, weights))
```

```python
model = CatBoostClassifier(eval_metric='AUC', early_stopping_rounds=10, auto_class_weights='Balanced', use_best_model=True, task_type="GPU") # custom_metric=['AUC:hints=skip_train~false']
```

```python
# train with best found params
model1 = CatBoostClassifier(iterations=2500,
                           learning_rate=0.1,
                           depth=7,
                           l2_leaf_reg=7,
                           eval_metric='AUC',
                            use_best_model=True,
                            task_type="GPU"
                           )

# train with best found params + auto weights
model2 = CatBoostClassifier(iterations=2500,
                           learning_rate=0.1,
                           depth=7,
                           l2_leaf_reg=7,
                           eval_metric='AUC',
                           auto_class_weights='Balanced',
                            use_best_model=True,
                            task_type="GPU"
                           )
```

```python
model3 = CatBoostClassifier(iterations=2500,
                           learning_rate=0.1,
                           depth=7,
                           l2_leaf_reg=7,
                           eval_metric='AUC',
                            class_weights=class_weights,
                            use_best_model=True,
                            task_type="GPU"
                           )
```

```python
#fit
train_pool = Pool(X_train, y_train, cat_features)
model1.fit(train_pool, eval_set=eval_dataset, verbose=False)
model2.fit(train_pool, eval_set=eval_dataset, verbose=False)
```
